```python
import numpy as np  # Импорт библиотеки для работы с массивами и векторной алгеброй
# from sklearn.metrics import accuracy_score  # Импорт функции для оценки точности модели
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt  # Импорт библиотеки для визуализации данных

# Настройка для отображения графиков непосредственно в блокноте (например, в Jupyter)
%matplotlib inline

import tensorflow as tf  # Импорт библиотеки TensorFlow для создания и обучения нейронных сетей
print("We're using TF", tf.__version__)  # Вывод версии TensorFlow
import tensorflow.keras as keras  # Импорт высокоуровневого API Keras для TensorFlow
import tensorflow.keras.models as M  # Импорт функционала для создания моделей в Keras
import tensorflow.keras.layers as L  # Импорт слоев для нейронных сетей в Keras
import tensorflow.keras.backend as K  # Импорт функций для низкоуровневой работы с тензорами и графами в Keras
import pandas as pd  # Импорт библиотеки для работы с данными в табличном виде (DataFrame)

import plotly.express as px
```

```python
# data = pd.read_csv("data1.csv")
# # data.head()

# points = data[['X', 'Y', 'Z']]
# points['X'] = -points['X']
# temp = points['Y']
# points['Y'] = points['Z']
# points['Z'] = temp

# points_next = points.drop(0)
# points_next.loc[points_next.index[-1]+1] = points.iloc[-1]
# points.reset_index(inplace=True, drop=True)
# points_next.reset_index(inplace=True, drop=True)
a1 = -2*np.pi/3
a2 = 2*np.pi/3
R = 3
z0 = 0
za = 1

t = np.linspace(0,15*np.pi,3000)
ind = np.linspace(0,2999,3000,dtype=int)

x = (R*(1-0.3*np.cos(2*t))*np.cos(a1 + (a2-a1)/(15*np.pi)*t) + np.random.rand(3000)/10).reshape(-1,1)
y = (R*(1-0.3*np.cos(2*t))*np.sin(a1 + (a2-a1)/(15*np.pi)*t) + np.random.rand(3000)/10).reshape(-1,1)
z = (z0 + za*np.cos(t) + np.random.rand(3000)/10).reshape(-1,1)

d = np.concatenate([x,y,z],axis=1)
data = pd.DataFrame(data=d,index=ind,columns=['X', 'Y', 'Z'])
data.to_csv("data.csv")

y = data.drop(0)
y.loc[y.index[-1]+1] = data.iloc[-1]
```

```python
fig = px.line_3d(data, x="X", y="Y", z="Z")
# fig = px.scatter_3d(data, x="X", y="Y", z="Z")
fig.show()
```

```python
data_train = data[0:2500]
data_test = data[2500:3000]

y_train = y[0:2500]
y_test = y[2500:3000]
```

```python
print(data_train.shape, y_train.shape)
print(data_test.shape, y_test.shape)
```

```python
K.clear_session()

model = M.Sequential()

model.add(L.Dense(6, activation='relu',input_shape=(data_train.shape[1],)))
# model.add(L.Dropout(0.3))
model.add(L.Dense(6, activation='relu'))
model.add(L.Dense(6, activation='relu'))


model.add(L.Dense(3, activation='linear'))
model.summary()
```

```python
model.compile(
    loss='mean_squared_error',
    optimizer='adam',
    metrics=['mean_squared_error']
)
```

```python
model.fit(
    data_train,
    y_train,
    batch_size=16,
    epochs=30,
    # validation_data=(X_test_float, y_test_oh)
)
```

```python
model.evaluate(data_test, y_test)
```

```python
predict_train = model.predict(data_train)
predict_test = model.predict(data_test)
```

```python
for_pred = data_test.iloc[0:1]
for i in range(500):
    temp = model.predict(for_pred.iloc[i:i+1])
    for_pred = pd.concat([for_pred,pd.DataFrame(data=temp,index=[2501+i],columns=['X','Y','Z'])],axis=0)

for_pred = for_pred.rename(columns={"X": "X_next", "Y": "Y_next", "Z": "Z_next"})
```
