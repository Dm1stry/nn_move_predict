```python
import numpy as np 
import pandas as pd  
import plotly.express as px
```

```python
a1 = -2*np.pi/3
a2 = 2*np.pi/3
R = 3
z0 = 0
za = 1
```

```python
def sin_gen(alpha_bounds, Radius, z_bounds, N, time_bounds, flex, noise_level, ):

    t = np.linspace(time_bounds[0], time_bounds[1], N)
    ind = (np.linspace(0, N-1, N, dtype=int) + (2*np.random.rand(N)-1)*noise_level).reshape(-1,1)

    x = (Radius*(1-flex*np.cos(2*t))*np.cos(alpha_bounds[0] + (alpha_bounds[1]-alpha_bounds[0])/(time_bounds[1])*t) + (2*np.random.rand(N)-1)*noise_level).reshape(-1,1)
    y = (Radius*(1-flex*np.cos(2*t))*np.sin(alpha_bounds[0] + (alpha_bounds[1]-alpha_bounds[0])/(time_bounds[1])*t) + (2*np.random.rand(N)-1)*noise_level).reshape(-1,1)
    z = (z_bounds[0] + z_bounds[1]*np.cos(t) + (2*np.random.rand(N)-1)*noise_level).reshape(-1,1)

    d = np.concatenate([ind,x,y,z],axis=1)
    data = pd.DataFrame(data=d,columns=['Time', 'X', 'Y', 'Z'])
    return data
```

```python
def show_sin(data, xyz = ["X","Y","Z"], markers = 0):
    fig = px.line_3d(data, x=xyz[0], y=xyz[1], z=xyz[2], markers=markers)
    fig.show()
```

```python
params1 = {'alpha_bounds': [-2*np.pi/3, 2*np.pi/3], 'Radius': 3, 'z_bounds': [0,2], 'N': 3000, 'time_bounds':[0, 15*np.pi], 'flex':0.3, 'noise_level': 0.05}
```

```python
data1 = sin_gen(**params1)
```

```python
show_sin(data=data1)
```
